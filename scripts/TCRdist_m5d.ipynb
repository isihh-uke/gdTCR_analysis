```python
clone_df, Z, distance_matrix = td.cdr3_tcrdist_clustering(td.m5d_donoragg_aaTCRdist3, 
                                                                    v_gene='TRDV2', 
                                                                    j_gene=['TRDJ1', 'TRDJ2', 'TRDJ3'], 
                                                                    length=[14, 15, 16, 17, 18, 19, 20, 21, 22],
                                                                    donor_category=None, shuffle=False, 
                                                                    method='complete',
                                                                    n_clusters=20, 
                                                                    criterion='maxclust', 
                                                                    n_neighbors=100,
                                                                    min_dist=0.3, n_components=2, 
                                                                    reduction='umap', 
                                                                    metric='euclidean', 
                                                                    title='m5d_20220602',
                                                                    n_jobs=128, 
                                                                    random_state=1)
```
